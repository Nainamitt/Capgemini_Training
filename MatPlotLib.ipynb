```python
# Question 3


import matplotlib.pyplot as plt
```

```python
# Scenario Question 
# A college is tracking student attendance across the first four months of the semester:

months = ['January' , 'Febuary' , 'March' , 'April']
attendance = [85 , 90 , 95 , 88]

# Plot line chart
plt.plot(months, attendance, marker='p', color='purple', linewidth=2)

# Labels and title
plt.xlabel("Months")
plt.ylabel("Average Students Present")
plt.title("Monthly Student Attendance Trend")

plt.grid(True, axis='y', alpha=0.4)
plt.show()
```

```python
 # Interpret the chart:
# - Which month had the highest attendance? 
# March had the highest attendance with 95 students.



# - Did attendance steadily increase, or were there ups and downs? 
# Attendance increased from January → February → March. Then , There was a drop in April. So, the trend was not perfectly steady.



# - What might explain the dip in April?
# Semester break or holidays or Increased assignments reducing class attendance
```
