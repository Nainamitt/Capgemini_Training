```python
# SCENARIO BASED QUESTIONS

# Scenario : FARM CROP YIELD ANALYSIS
# 3 CROPS - WHEAT , RICE , CORN 
# 4 SEASON - SPRING , SUMMER , AUTUMN , WINTER 

import numpy as np
```

```python
# Create a 2D array representing crop yields for 3 crops across 4 seasons.
# - Rows = crops
# - Columns = seasons
    
matrix = np.array([[10,233,23,23] , [34,34,524,65] , [34,645,76,87]])
matrix
```

```python
# Increase all yields by 10% to account for improved irrigation.
updated = np.array(matrix * 0.1)
updated
```

```python
# Extract only Wheatâ€™s yields (the first crop).
updated[0,:]
```

```python
# Find the average yield per crop across all seasons.
avg = updated.mean(axis=1)
avg
```

```python
 # Identify yields greater than 3000 kg
updated[updated>50]
```
