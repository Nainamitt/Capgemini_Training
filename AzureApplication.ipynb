```python
from dotenv import load_dotenv
import os

# Load env variables
load_dotenv(dotenv_path=r"C:\Users\naina\Capgemini\Generative Ai\.env")

api_key = os.getenv("OPENAI_API_KEY")
endpoint = os.getenv("OPENAI_BASE_URL")
deployment_name = os.getenv("DEPLOYMENT_NAME")
```

```python
# Scenario:
# You are a data science trainer preparing a demo for a group of first-time managers in e-commerce.
# You want to show them how to securely connect to Azure OpenAI from Google Colab, send a simple query,
#  and interpret the response. The demo emphasizes security (using secrets) and clarity (structured output).
```

```python
from openai import AzureOpenAI

# Stop if missing
if not api_key or not endpoint or not deployment_name:
    raise ValueError("Missing environment variables!")

# Initialize client
client = AzureOpenAI(
    api_key=api_key,
    api_version="2024-02-15-preview",
    azure_endpoint=endpoint
)

# API call
response = client.chat.completions.create(
    model=deployment_name,
    messages=[
        {"role": "user", "content": "What is the capital of France?"}
    ]
)

print("\nResponse:\n", response.choices[0].message.content)
```

```python
from openai import AzureOpenAI

# Initialize Azure client
client = AzureOpenAI(
    api_key=api_key,
    api_version="2024-02-15-preview",
    azure_endpoint=endpoint
)
```

```python
def healthcare_ai_assistant(user_input):
    response = client.chat.completions.create(
        model=deployment_name,
        messages=[
            {
                "role": "system",
                "content": """
                You are a Healthcare AI Training Assistant.

                Your role:
                - Explain AI concepts in healthcare in simple terms
                - Provide real-world healthcare examples
                - Be clear, ethical, and safe
                - Avoid giving medical advice (only educational content)
                """
            },
            {
                "role": "user",
                "content": user_input
            }
        ]
    )
    
    return response.choices[0].message.content
```

```python
print(healthcare_ai_assistant("How is AI used in disease prediction?"))
```
