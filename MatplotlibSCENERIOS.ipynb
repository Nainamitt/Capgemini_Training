```python
# Create a figure of size 85 inches and give me an axis (ax) to draw my chart on.
fig , ax = plt.subplots(figsize = (8,5))

months = ['JAN' , 'FEB' , 'MARCH' , 'APRIL' , 'MAY' , 'JUNE']
sales = [42000 , 34000 , 45000 , 65000 , 54000 , 45000]

# ax.plot(months , sales , color = "darkblue" , linewidth = 2.5 , marker = 'o' , markersize = 10)
ax.plot(months , sales , color = "darkblue" , linewidth = 2.5 , marker = 'p' , markersize = 10)


ax.set_title('Monthly Sales 2024' , fontsize = 14)
ax.set_xlabel("Month")
ax.set_ylabel("Revenue ($)")
ax.grid(axis = 'y' , alpha = 0.4)
# Fix the spacing of the plot so everything fits nicely inside the figure
plt.tight_layout()
plt.show()
```

```python
ad_spend = np.array([2000 , 4000 , 6000 , 8000 , 10000 , 12000 , 14000 , 16000 , 18000])
sales = np.array([7000 , 15000 , 21000 , 25000 , 32000 , 36000 , 42000 , 47000 , 52000])

plt.scatter(ad_spend , sales , color = "darkblue" ,  alpha = 0.7)
z = np.polyfit(ad_spend , sales , 1)
p = np.poly1d(z)
plt.plot(ad_spend , p(ad_spend), 'r--' , label = "Trend")

plt.xlabel("Ad Spend ($)")
plt.ylabel("Sales ($)")
plt.title("Add Spend VS Sales")
# shows label Trend
plt.legend()
plt.show()
```

```python
import random
```

```python
heads = 0
total = 1000

for i in range(total):
    result = random.randint(0,1)
    if result == 1:
        heads = heads+1

tails = total - heads

print(tails)
print(heads)
print("Probability : " , tails/total)
```

```python
heights = np.random.normal(165 , 8 , 200)
print("Shortest" , round(heights.min() , 1))
print("Tallest" , round(heights.max() , 1))
print("Average" , round(heights.mean() , 1))

plt.hist(heights , bins = 20 , color = "pink" , edgecolor = "white")
plt.axvline(heights.mean() , color = "red" , linewidth  = 2 , label = "Average")
plt.title("Heights of 200 students")
plt.xlabel("Heights (cm)")
plt.ylabel("Number of Students")
plt.legend()
plt.show()
```

```python
from scipy.stats import binom
```

```python
# Probability of getting exactly 3 successes in 10 trials when success probability is 0.25.
# PMF Gives probability of getting exactly k successes 

n = 10
p = 0.25

p3 = binom.pmf(3,n,p)
# ROUND HUME 3 DECIMAL POINT TK ANSWER DETA HAI
print("P(exactly 3) = " , round(p3 , 3))
```

```python
# AT LEAST 5 CORRECT- 5 OR 5 SAI JADHA TOH 4 KI NIKALO USSKO -1
# CDF HUM USE KARTE HAI JAB Gives probability of getting  k successes 

p5 = 1 - binom.cdf(4,n,p)
print("P(5 or more) = " , round(p5 , 3))
```
