```python
# LAB 1: Python Fundamentals + File Handling

# 🎯 Objective:
# Build a Student Result Processing System

# 🧩 Tasks:
# Take input for 5 students (name + marks)

# Calculate percentage & grade

# Store results in a .txt file

# Read file and display toppers

# 📤 Expected Output:
# Structured file output

# Sorted results
```

```python
students = []

for i in range(5):
    name = input("Enter name: ")
    marks = float(input("Enter marks: "))
    percent = marks
    grade = "A" if percent >= 90 else "B" if percent >= 75 else "C"
    students.append((name, percent, grade))

```

```python
# Write to file
with open("results.txt", "w") as f:
    for s in students:
        f.write(f"{s[0]}, {s[1]}, {s[2]}\n")
```

```python
# Read + find topper
with open("results.txt", "r") as f:
    data = [line.strip().split(", ") for line in f]

data.sort(key=lambda x: float(x[1]), reverse=True)

print("Topper:", data[0])
```

```python
# LAB 2: Data Types + Control Flow


# 🎯 Objective:
# Build a Data Analyzer Tool

# 🧩 Tasks:
# Input a mixed list

# Separate integers, floats, strings

# Count frequency of each type

# Print summary using loops
```

```python
data = [10, 2.5, "Anuu", 10, "Naina", 3.5 , "MoM"]

ints, floats, strings = [], [], []

for i in data:
    if isinstance(i, int):
        ints.append(i)
    elif isinstance(i, float):
        floats.append(i)
    else:
        strings.append(i)
```

```python
print("Integers:", ints)
print("Floats:", floats)
print("Strings:", strings)
```

```python
print("Count:", len(ints), len(floats), len(strings))
```

```python
# LAB 3: Functions + Recursion


# 🎯 Objective:
# Create a Math Utility Module

# 🧩 Tasks:
# Factorial using recursion

# Fibonacci using recursion

# Lambda for square & cube

# Compare execution time
```

```python
import time

def fact(n):
    return 1 if n == 0 else n * fact(n-1)

def fib(n):
    return n if n <= 1 else fib(n-1) + fib(n-2)

square = lambda x: x*x
cube = lambda x: x*x*x
```

```python
start = time.time()
print(fact(5), fib(5))
print("Time:", time.time() - start)
```

```python
# LAB 4: Regex + File Processing


# 🎯 Objective:
# Build an Email Extractor Tool

# 🧩 Tasks:
# Read a text file

# Extract valid emails using regex

# Count unique emails

# Save results in CSV
```

```python
import re
import csv
import os

if not os.path.exists("data.txt"):
    print("File not found! Creating sample file...")
    with open("data.txt", "w") as f:
        f.write("test@gmail.com\nuser@yahoo.com")
        
with open("data.txt", "r") as f:
    text = f.read()

emails = re.findall(r'\b[\w.-]+@[\w.-]+\.\w+\b', text)
unique_emails = list(set(emails))

with open("emails.csv", "w", newline="") as f:
    writer = csv.writer(f)
    for email in unique_emails:
        writer.writerow([email])

print("Total unique emails:", len(unique_emails))
```

```python
#  LAB 5: Exception Handling

# 🎯 Objective:
# Develop a Robust Calculator

# 🧩 Tasks:
# Handle invalid input

# Handle division by zero

# Use try–except–finally

# Create custom exception
```

```python
class CustomError(Exception):
    pass

try:
    a = float(input("Enter a: "))
    b = float(input("Enter b: "))
    op = input("Enter operator: ")

    if op == "/" and b == 0:
        raise CustomError("Division by zero!")

    result = eval(f"{a}{op}{b}")
    print("Result:", result)

except CustomError as e:
    print(e)
except Exception:
    print("Invalid input!")
finally:
    print("Execution complete")
```

```python
# LAB 6: OOP + Inheritance

# 🎯 Objective:
# Design an Employee Management System

# 🧩 Tasks:
# Base class: Employee

# Derived: Manager, Developer

# Override salary method

# Display employee details
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def get_salary(self):
        return self.salary

class Manager(Employee):
    def get_salary(self):
        return self.salary + 5000

class Developer(Employee):
    def get_salary(self):
        return self.salary + 3000

e1 = Manager("A", 50000)
e2 = Developer("B", 40000)

print(e1.name, e1.get_salary())
print(e2.name, e2.get_salary())
```

```python
# LAB 7: Decorators + Generators

# 🎯 Objective:
# Optimize performance using advanced Python

# 🧩 Tasks:
# Create decorator for logging

# Create generator for large data

# Compare memory usage
```

```python
import sys

def logger(func):
    def wrapper(*args):
        print("Executing:", func.__name__)
        return func(*args)
    return wrapper

@logger
def test():
    print("Hello")

def generator(n):
    for i in range(n):
        yield i

gen = generator(1000000)
print(sys.getsizeof(gen))  # memory efficient
```

```python
#  LAB 8: Multithreading

# 🎯 Objective:
# Simulate Parallel File Downloader

# 🧩 Tasks:
# Download multiple files using threads

# Measure execution time

# Compare with sequential execution
```

```python
import threading
import time

def task(n):
    print(f"Downloading {n}")
    time.sleep(2)

start = time.time()

threads = []
for i in range(3):
    t = threading.Thread(target=task, args=(i,))
    t.start()
    threads.append(t)

for t in threads:
    t.join()

print("Time:", time.time() - start)
```

```python
# LAB 9: NumPy + Pandas

# 🎯 Objective:
# Perform Data Analysis

# 🧩 Tasks:
# Load dataset

# Handle missing values

# Perform filtering

# Compute mean, median, aggregation
```

```python
import pandas as pd

df = pd.read_csv("Student Success & Career Path.csv")

df.fillna(df.mean(numeric_only=True), inplace=True)

print(df[df['Final_Exam_Score'] > 50])

print("Mean:", df['Final_Exam_Score'].mean())
print("Median:", df['Final_Exam_Score'].median())
```

```python
# LAB 10: Mini Project (FINAL)
# 🎯 Objective:
# Build a Student Management System (End-to-End)

# 🧩 Features:
# Add student

# Search student

# Store in CSV

# Display analytics (topper, avg marks)
```

```python
import csv

def add_student():
    with open("students.csv", "a", newline="") as f:
        writer = csv.writer(f)
        name = input("Name: ")
        marks = float(input("Marks: "))
        writer.writerow([name, marks])

def display():
    with open("students.csv", "r") as f:
        data = list(csv.reader(f))
        data.sort(key=lambda x: float(x[1]), reverse=True)

        print("Topper:", data[0])
        avg = sum(float(x[1]) for x in data) / len(data)
        print("Average:", avg)

while True:
    ch = input("1.Add 2.Display 3.Exit: ")
    if ch == "1":
        add_student()
    elif ch == "2":
        display()
    else:
        break
```
