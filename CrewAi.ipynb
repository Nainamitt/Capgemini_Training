```python
# Scenario: AI Content Automation for a Startup
# 🏢 Problem Statement

# A startup wants to:

# Stay updated with latest AI trends

# Automatically create LinkedIn posts

# Save time on manual research & writing

# 👉 So we build a multi-agent AI system to automate this using CrewAI.
```

```python
import os
os.environ["CREWAI_TRACING_ENABLED"] = "true"
from crewai import Agent, Task, Crew
from dotenv import load_dotenv
from langchain_litellm import ChatLiteLLM # Updated import

load_dotenv()

# -----------------------------
# 1. Set Groq API Key
# -----------------------------
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

# ❌ REMOVE any OPENAI key completely
if "OPENAI_API_KEY" in os.environ:
    del os.environ["OPENAI_API_KEY"]

# -----------------------------
# 2. Define LLM (using ChatLiteLLM for Groq)
# -----------------------------
llm = ChatLiteLLM(model="groq/llama-3.3-70b-versatile")

# -----------------------------
# 3. Define Agents
# -----------------------------
researcher = Agent(
    role="Researcher",
    goal="Find latest AI news",
    backstory="Expert in AI trends",
    llm=llm, # Use the explicitly defined llm object
    verbose=True
)

writer = Agent(
    role="Writer",
    goal="Write LinkedIn post",
    backstory="Expert content creator",
    llm=llm, # Use the explicitly defined llm object
    verbose=True
)

# -----------------------------
# 4. Tasks
# -----------------------------
research_task = Task(
    description="Find 5 latest AI news trends",
    expected_output="List with summaries",
    agent=researcher
)

write_task = Task(
    description="Write LinkedIn post",
    expected_output="Final post",
    agent=writer,
    context=[research_task]
)

# -----------------------------
# 5. Crew
# -----------------------------
crew = Crew(
    agents=[researcher, writer],
    tasks=[research_task, write_task],
    verbose=True
)

# -----------------------------
# 6. Run
# -----------------------------
result = crew.kickoff()

print("\n✅ FINAL OUTPUT:\n")
print(result)
```

```python
# Scenario: Automated Learning Hub for an EdTech Company
# 🎓 Problem Statement
# An EdTech company wants to:
# - Keep track of the latest educational technology innovations
# - Automatically generate weekly blog posts for their website
# - Reduce time spent on manual research and content drafting
# 👉 So we build a multi-agent AI system using CrewAI to automate this workflow:
# - Trend Monitoring Agent → Continuously scans for new EdTech tools, teaching strategies, and AI-in-education updates
# - Content Summarization Agent → Condenses research into digestible insights tailored for educators and learners
# - Blog Writing Agent → Crafts engaging, SEO-friendly blog posts automatically
# - Publishing Agent → Schedules and posts content directly to the company’s website
```

```python
import os
os.environ["CREWAI_TRACING_ENABLED"] = "true"
from dotenv import load_dotenv
from crewai import Agent, Task, Crew

from langchain_litellm import ChatLiteLLM
load_dotenv()


# -----------------------------
# 1. API Setup (Groq)
# -----------------------------
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

if "OPENAI_API_KEY" in os.environ:
    del os.environ["OPENAI_API_KEY"]

# -----------------------------
# 2. LLM
# -----------------------------
llm = ChatLiteLLM(model="groq/llama-3.3-70b-versatile")

# -----------------------------
# 3. Agents
# -----------------------------

# 🔍 Trend Monitoring Agent
trend_agent = Agent(
    role="Trend Monitoring Specialist",
    goal="Track latest EdTech innovations, AI in education, and teaching strategies",
    backstory="Expert in analyzing global education trends and emerging technologies",
    llm=llm,
    verbose=True
)

# 🧠 Content Summarization Agent
summary_agent = Agent(
    role="Content Summarizer",
    goal="Convert research into short, easy-to-understand insights",
    backstory="Specialist in simplifying complex educational content",
    llm=llm,
    verbose=True
)

# ✍️ Blog Writing Agent
writer_agent = Agent(
    role="Blog Writer",
    goal="Write engaging and SEO-friendly blog posts for an EdTech website",
    backstory="Professional content writer with expertise in education blogs",
    llm=llm,
    verbose=True
)

# 📤 Publishing Agent
publish_agent = Agent(
    role="Publishing Manager",
    goal="Format, schedule, and publish blog posts",
    backstory="Expert in content publishing and CMS platforms",
    llm=llm,
    verbose=True
)

# -----------------------------
# 4. Tasks
# -----------------------------

# Task 1: Trend Research
trend_task = Task(
    description="Find latest 5 EdTech trends, AI in education updates, and new teaching tools",
    expected_output="Bullet list of trends with short explanations",
    agent=trend_agent
)

# Task 2: Summarization
summary_task = Task(
    description="Summarize the research into simple insights for students and educators",
    expected_output="Concise and clear summary",
    agent=summary_agent,
    context=[trend_task]
)

# Task 3: Blog Writing
blog_task = Task(
    description="Write a detailed, engaging, SEO-friendly weekly blog post based on the summary",
    expected_output="Full blog article with title, headings, and conclusion",
    agent=writer_agent,
    context=[summary_task]
)

# Task 4: Publishing
publish_task = Task(
    description="Format the blog post properly and prepare it for publishing (add headings, metadata, schedule)",
    expected_output="Final publish-ready blog content",
    agent=publish_agent,
    context=[blog_task]
)

# -----------------------------
# 5. Crew
# -----------------------------
crew = Crew(
    agents=[trend_agent, summary_agent, writer_agent, publish_agent],
    tasks=[trend_task, summary_task, blog_task, publish_task],
    verbose=True
)

# -----------------------------
# 6. Run
# -----------------------------
result = crew.kickoff()

print("\n✅ FINAL OUTPUT:\n")
print(result)
```

```python
# Scenario: Smart Job Application Assistant
# 🏢 Problem Statement

# Students struggle to:

# Analyze job descriptions

# Match their skills

# Write tailored resumes & cover letters

# 👉 So we build a multi-agent AI system that automates job application preparation.

# 🤖 System Overview
# 🔹 Agent 1: Job Analyzer

# Reads job description

# Extracts required skills

# 🔹 Agent 2: Resume Optimizer

# Matches candidate skills with job

# Suggests improvements

# 🔹 Agent 3: Cover Letter Writer

# Generates a personalized cover letter

# 🔄 Workflow
# Job Description → Skill Analysis → Resume Optimization → Cover Letter Generation


import os
from crewai import Agent, Task, Crew

from dotenv import load_dotenv

load_dotenv()
# -----------------------------
# 1. API Key Setup
# -----------------------------
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

# -----------------------------
# 2. Define Agents
# -----------------------------
job_analyzer = Agent(
    role="Job Analyzer",
    goal="Extract key skills and requirements from job descriptions",
    backstory=(
        "You are an HR expert who understands job postings and identifies "
        "important skills, qualifications, and expectations."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

resume_optimizer = Agent(
    role="Resume Optimizer",
    goal="Align candidate resume with job requirements",
    backstory=(
        "You are a career coach who improves resumes by matching them "
        "with job descriptions and highlighting relevant skills."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

cover_letter_writer = Agent(
    role="Cover Letter Writer",
    goal="Write a personalized cover letter",
    backstory=(
        "You are a professional writer who crafts compelling cover letters "
        "tailored to specific job roles."
    ),
    llm="groq/llama-3.3-70b-versatile",
    verbose=True
)

# -----------------------------
# 3. Input Data (Scenario)
# -----------------------------
job_description = """
We are looking for a Data Analyst with experience in Python, SQL, Power BI,
and data visualization. The candidate should be able to analyze datasets,
create dashboards, and communicate insights effectively.
"""

candidate_resume = """
I am a BTech student skilled in Python and basic SQL.
I have created Power BI dashboards and completed projects on data analysis.
"""

# -----------------------------
# 4. Tasks
# -----------------------------
analysis_task = Task(
    description=f"""
    Analyze the following job description and extract:
    - Required skills
    - Key responsibilities
    - Tools/technologies

    Job Description:
    {job_description}
    """,
    expected_output="Structured list of skills, responsibilities, and tools",
    agent=job_analyzer
)

resume_task = Task(
    description=f"""
    Compare the candidate resume with the job requirements and:
    - Identify skill gaps
    - Suggest improvements
    - Optimize the resume

    Resume:
    {candidate_resume}
    """,
    expected_output="Improved resume with suggestions",
    agent=resume_optimizer,
    context=[analysis_task]
)

cover_letter_task = Task(
    description="""
    Create a personalized cover letter using:
    - Job requirements
    - Candidate resume
    - Improvements suggested

    Make it professional and engaging.
    """,
    expected_output="Final cover letter",
    agent=cover_letter_writer,
    context=[analysis_task, resume_task]
)

# -----------------------------
# 5. Crew Setup
# -----------------------------
crew = Crew(
    agents=[job_analyzer, resume_optimizer, cover_letter_writer],
    tasks=[analysis_task, resume_task, cover_letter_task],
    verbose=True,
    tracing=True   # good for demo
)

# -----------------------------
# 6. Run
# -----------------------------
result = crew.kickoff()

print("\n🚀 FINAL OUTPUT:\n")
print(result)
```

```python
# Problem Statement
# Enterprise sales teams often struggle to:
# - Analyze client requirements from RFPs (Requests for Proposals)
# - Match company offerings to client needs
# - Draft tailored proposals that stand out
# - Ensure consistency across documents
# 👉 So we build a multi-agent AI system that automates sales proposal preparation.

# 🤖 System Overview
# - 🔹 Agent 1: RFP Analyzer
# - Reads client’s request for proposal
# - Extracts key requirements, budget constraints, and evaluation criteria
# - 🔹 Agent 2: Solution Mapper
# - Matches company products/services to client needs
# - Suggests tailored value propositions and differentiators
# - 🔹 Agent 3: Proposal Writer
# - Generates a polished, client-specific proposal
# - Ensures alignment with brand tone and compliance standards

# 🔄 Workflow
# RFP Document → Requirement Analysis → Solution Mapping → Proposal Generation
```

```python

import os
from crewai import Agent, Task, Crew
from dotenv import load_dotenv
from langchain_litellm import ChatLiteLLM
load_dotenv()

# -----------------------------
# 1. API Setup (Groq)
# -----------------------------
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

# Remove OpenAI if present
if "OPENAI_API_KEY" in os.environ:
    del os.environ["OPENAI_API_KEY"]

# -----------------------------
# 2. LLM
# -----------------------------
llm = ChatLiteLLM(model="groq/llama-3.3-70b-versatile")

# -----------------------------
# 3. Agents
# -----------------------------

# 🔹 RFP Analyzer
rfp_analyzer = Agent(
    role="RFP Analyzer",
    goal="Extract client requirements, budget, and evaluation criteria from RFP",
    backstory="Expert in analyzing enterprise RFP documents and identifying key business needs",
    llm=llm,
    verbose=True
)

# 🔹 Solution Mapper
solution_mapper = Agent(
    role="Solution Mapper",
    goal="Map company solutions to client requirements",
    backstory="Experienced sales strategist who aligns products with client needs",
    llm=llm,
    verbose=True
)

# 🔹 Proposal Writer
proposal_writer = Agent(
    role="Proposal Writer",
    goal="Create professional and persuasive business proposals",
    backstory="Expert business writer specializing in client proposals and enterprise sales",
    llm=llm,
    verbose=True
)

# -----------------------------
# 4. Input Data (Sample RFP)
# -----------------------------
rfp_text = """
We are seeking a cloud-based data analytics solution for our enterprise.
The solution should support real-time analytics, dashboard creation, and integration with existing systems.

Requirements:
- Data visualization dashboards
- Integration with APIs
- Scalable cloud infrastructure
- Security compliance

Budget: $50,000
Evaluation Criteria:
- Cost efficiency
- Scalability
- Ease of use
"""

company_profile = """
Our company provides cloud-based analytics platforms with real-time dashboards,
API integrations, and enterprise-grade security. We specialize in scalable solutions
using AWS and Power BI.
"""

# -----------------------------
# 5. Tasks
# -----------------------------

# Task 1: RFP Analysis
analysis_task = Task(
    description=f"""
    Analyze the given RFP and extract:

    - Client requirements
    - Budget constraints
    - Evaluation criteria
    - Key objectives

    Return the output in bullet points.

    RFP Document:
    {rfp_text}
    """,
    expected_output="Structured RFP analysis",
    agent=rfp_analyzer
)

# Task 2: Solution Mapping
solution_task = Task(
    description=f"""
    Based on the RFP analysis and company profile:

    - Match company offerings with client requirements
    - Suggest suitable solutions
    - Highlight value propositions
    - Mention differentiators

    Company Profile:
    {company_profile}
    """,
    expected_output="Mapped solutions with value propositions",
    agent=solution_mapper,
    context=[analysis_task]
)

# Task 3: Proposal Writing
proposal_task = Task(
    description="""
    Create a professional business proposal using:

    - Client requirements
    - Suggested solutions

    Include:
    - Introduction
    - Understanding of client needs
    - Proposed solution
    - Benefits and value proposition
    - Conclusion

    Keep tone formal and persuasive.
    """,
    expected_output="Final business proposal document",
    agent=proposal_writer,
    context=[analysis_task, solution_task]
)

# -----------------------------
# 6. Crew Setup
# -----------------------------
crew = Crew(
    agents=[rfp_analyzer, solution_mapper, proposal_writer],
    tasks=[analysis_task, solution_task, proposal_task],
    verbose=True,
    tracing=True
)

# -----------------------------
# 7. Run
# -----------------------------
result = crew.kickoff()

print("\n🚀 FINAL OUTPUT:\n")
print(result)
```
