```python
# 📘 Scenario: Smart Manufacturing Insights
# Context:
# You are part of a data science team in a smart manufacturing company. Your manager wants you to explore how different 
# technical concepts (like maintenance, telemetry, quality control, supply chain) are related in company documents. 
# By training a small Word2Vec model, you can uncover semantic similarities between words and measure how closely they are
#  connected. This helps in building recommendation systems, knowledge graphs, or even automated reporting tools.
# Task:
# Given a short corpus describing manufacturing operations, train a Word2Vec model and analyze word similarities
```

```python
import nltk
```

```python
nltk.download("punkt")
nltk.download("punkt_tab")
```

```python
# splits text into sentences -- sent_tokenize
# splits sentences into words -- word_tokenize

from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
# which converts words into numerical vectors.

from gensim.models import Word2Vec
```

```python
corpus = """
Manufacturing relies on predictive maintenance and supply chain optimization.
Data engineers build pipelines, while analysts monitor KPIs and anomalies.
Robotics and IoT sensors stream telemetry to cloud databases for real-time insights.
Quality control uses computer vision to detect defects on the shop floor.
"""
```

```python
sentences = [word_tokenize(s.lower()) for s in sent_tokenize(corpus)]
```

```python
model = Word2Vec(
    sentences,
    vector_size=50,
    window=5,
    min_count=1,
    workers=2,
    sg=1
)
```

```python
for target in ["maintenance", "supply", "quality", "telemetry"]:
    print(f"\nTop similar to '{target}':")
    try:
        for w , score in model.wv.most_similar(target, topn=5):
             print(f"{w:15} -> {score:.3f}")
            
    except KeyError:
        print("Word not in vocabulary.")
```

```python
from numpy import dot
from numpy.linalg import norm

def cosine(u, v):
    return dot(u, v) / (norm(u) * norm(v))

pairs = [("maintenance", "telemetry"),
         ("quality", "defects"),
         ("supply", "optimization")]
```

```python
print("\nCosine similarities:")
for a, b in pairs:
    try:
        sim = cosine(model.wv[a], model.wv[b])
        print(f"{a:12} ~ {b:12} -> {sim:.3f}")
    except KeyError:
        print(f"Missing word: {a} or {b}")
```

```python
# 📘 Scenario: Healthcare Research Insights
# Context:
# You are part of a data science team in a healthcare research institute. 
# Your manager wants you to analyze medical reports and research papers to explore how different technical concepts (like diagnosis, treatment, symptoms, patients, medication) are related.
# By training a small Word2Vec model, you can uncover semantic similarities between words and measure how closely they are connected. This helps in building clinical decision support systems, medical knowledge graphs, or even automated literature reviews.
# Task:
# Given a short corpus describing healthcare operations, train a Word2Vec model and analyze word similarities.
```

```python
import nltk
nltk.download("punkt")

from nltk.tokenize import sent_tokenize, word_tokenize
from gensim.models import Word2Vec
from numpy import dot
from numpy.linalg import norm
```

```python
# Healthcare corpus
corpus = """
Doctors diagnose diseases using patient symptoms and medical history.
Treatment plans may include medication, therapy, or surgery.
Hospitals monitor patient recovery and adjust medication accordingly.
Researchers analyze clinical data to improve diagnosis and treatment methods.
Pharmaceutical companies develop new drugs to treat chronic diseases.
"""

# Sentence → word tokenization
sentences = [word_tokenize(s.lower()) for s in sent_tokenize(corpus)]
```

```python
sentences
```

```python
model = Word2Vec(
    sentences,
    vector_size=50,
    window=4,
    min_count=1,
    workers=2,
    sg=1
)
```

```python
model
```

```python
targets = ["diagnosis", "treatment", "patient", "medication"]

for word in targets:
    print(f"\nWords similar to '{word}':")
    try:
        for w, score in model.wv.most_similar(word, topn=5):
            print(f"{w:15} -> {score:.3f}")
    except KeyError:
        print("Word not found")
```

```python
def cosine(u, v):
    return dot(u, v) / (norm(u) * norm(v))

# Word pairs for comparison
pairs = [
    ("diagnosis", "symptoms"),
    ("treatment", "medication"),
    ("patient", "hospital")
]

print("\nCosine Similarities:")

for a, b in pairs:
    try:
        sim = cosine(model.wv[a], model.wv[b])
        print(f"{a:12} ~ {b:12} -> {sim:.3f}")
    except KeyError:
        print("Word missing")
```
