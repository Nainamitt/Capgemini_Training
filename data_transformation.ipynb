```python
# SCENARIO BASED QUESTION

# CORPORATE EMPLOYEEE  Performance Dashboard A company tracks quarterly performance scores of employees across different departments.
# The HR analytics team wants to analyze this dataset to identify top performers, departmental trends, and highlight exceptional scores

import pandas as pd
```

```python
df = pd.read_csv("dataSet.csv")
```

```python
ans = df[df["Score"] > 90]
ans
```

```python
df[df["Department"] == "Sales"]
```

```python
df.sort_values("Score" , ascending = False)
```

```python
df.sort_values(["Department" , "Score"] , ascending = [True , False])
```

```python
df.groupby("Department")["Score"].mean()
```

```python
df.groupby("Quarter")["Score"].max()
```
